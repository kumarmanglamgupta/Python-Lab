2.Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school. Also generate a horizontal bar chart based on the result and explain the conclusion.

 Input: student_data = pd.DataFrame({ 'school_code': ['s001','s002','s003','s001','s002','s004'],

 'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'], 

'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'], 

'age': [12, 12, 13, 13, 14, 12], 

'height': [173, 192, 186, 167, 151, 159], 

'weight': [35, 32, 33, 30, 31, 32], 

'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']}, )

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
student_data = pd.DataFrame({
    'school_code': ['s001', 's002', 's003', 's001', 's002', 's004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']
})

# Group by 'school_code' and calculate mean, min, and max age
age_stats = student_data.groupby('school_code')['age'].agg(['mean', 'min', 'max'])

# Rename columns for clarity
age_stats.rename(columns={'mean': 'mean_age', 'min': 'min_age', 'max': 'max_age'}, inplace=True)

# Print the aggregated statistics
print("Age Statistics by School Code:")
print(age_stats)

# Plot horizontal bar charts
plt.figure(figsize=(10, 6))

# Plot mean age
plt.subplot(1, 3, 1)
age_stats['mean_age'].plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Mean Age')
plt.xlabel('Age')
plt.ylabel('School Code')

# Plot minimum age
plt.subplot(1, 3, 2)
age_stats['min_age'].plot(kind='barh', color='lightgreen', edgecolor='black')
plt.title('Min Age')
plt.xlabel('Age')
plt.ylabel('School Code')

# Plot maximum age
plt.subplot(1, 3, 3)
age_stats['max_age'].plot(kind='barh', color='lightcoral', edgecolor='black')
plt.title('Max Age')
plt.xlabel('Age')
plt.ylabel('School Code')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
