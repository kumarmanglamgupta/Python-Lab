2.Suppose you're a sales manager for an e-commerce company, and you want to create a figure with subplots to compare the sales performance of different product categories over time. You have sales data for four product categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share your conclusion/analysis.

Input: months = np.arange(1, 13) 

electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000]) 

clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000]) 

home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])

sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
months = np.arange(1, 13)  # Months from January (1) to December (12)

# Sales data for each category
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])

# Create subplots with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True)

# Plot Electronics Sales
axs[0, 0].bar(months - 0.3, electronics_sales, width=0.3, color='blue', label='Electronics')
axs[0, 0].set_title('Electronics Sales')
axs[0, 0].set_ylabel('Sales ($)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot Clothing Sales
axs[0, 1].bar(months - 0.3, clothing_sales, width=0.3, color='green', label='Clothing')
axs[0, 1].set_title('Clothing Sales')
axs[0, 1].set_ylabel('Sales ($)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot Home & Garden Sales
axs[1, 0].bar(months - 0.3, home_garden_sales, width=0.3, color='red', label='Home & Garden')
axs[1, 0].set_title('Home & Garden Sales')
axs[1, 0].set_ylabel('Sales ($)')
axs[1, 0].set_xlabel('Month')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot Sports & Outdoors Sales
axs[1, 1].bar(months - 0.3, sports_outdoors_sales, width=0.3, color='purple', label='Sports & Outdoors')
axs[1, 1].set_title('Sports & Outdoors Sales')
axs[1, 1].set_ylabel('Sales ($)')
axs[1, 1].set_xlabel('Month')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Set common x-axis label
plt.xlabel('Month')
plt.tight_layout()  # Adjust layout to make room for labels
plt.show()
```

Conclusion:


1) Electronics Sales:

Trend: The sales of Electronics show a generally upward trend, with significant growth from January to December.

Insights: This indicates that Electronics is performing very well throughout the year. The increase in sales could be due to seasonal promotions, new product launches, or increasing demand.


2) Clothing Sales:

Trend: Clothing sales also exhibit a steady increase over the year. The growth is consistent, showing a gradual rise each month.

Insights: The steady growth suggests that the clothing category is performing reliably. This could be attributed to ongoing fashion trends or effective marketing strategies.


3) Home & Garden Sales:

Trend: Home & Garden sales show a steady upward trend similar to Clothing, with consistent growth over the year.

Insights: This indicates that the Home & Garden category is experiencing regular demand. The stability in growth suggests a consistent interest in these products, possibly driven by seasonal changes or home improvement trends.


4) Sports & Outdoors Sales:

Trend: Sports & Outdoors sales also demonstrate a steady increase, although the growth is slightly slower compared to Electronics and Clothing.

Insights: The slower but steady growth could indicate a more niche market. The incremental increases might be due to seasonal sports activities or gradual market penetration.
