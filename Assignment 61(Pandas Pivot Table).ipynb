```python
#importing libraries
import pandas as pd
import numpy as np
#Read in our sales data into our DataFrame
df = pd.read_csv("/Users/jitendragupta/Downloads/salesdata.csv")
df
```

```python
#creating a pivot table
print(pd.pivot_table(df,index=["Item"],values=['Units'],aggfunc=np.sum))
```
