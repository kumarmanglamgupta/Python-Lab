1. Create a pie chart to visualize the distribution of your monthly income by source. You have collected data on the various sources of your income, such as salary, freelance work, investments, and rental income. Share your conclusion/analysis.

 Input: income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other'] 

monthly_income = [5000, 1500, 1000, 600, 400]

```python
import matplotlib.pyplot as plt

# Data
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(8, 8))  # Set the size of the figure

# Create a pie chart
ax.pie(
    monthly_income,                # Data to plot
    labels=income_sources,         # Labels for each section
    autopct='%1.1f%%',             # Format for displaying percentage values
    startangle=140,                # Start angle of the pie chart for better visual appeal
    colors=plt.cm.Paired(range(len(income_sources)))  # Color map for different segments
)

# Add a title
ax.set_title('Distribution of Monthly Income by Source')

# Display the pie chart
plt.show()
```

Conclusion:


1) Primary Income Source: Your salary is the primary source of income, which forms the bulk of your monthly earnings.


2) Supplementary Income: Freelance work provides a significant additional source of income, which could indicate skills or services offered outside of your main job.


3) Diversification: You have diversified income sources through investments and rental properties, though these contribute less than your main sources.


4) Minor Sources: Income classified as "Other" represents a small part of your total income, suggesting occasional or irregular contributions.
